# 2) Minimum widths and alignment between columns

```
# create a list of tuples
marvel = [('Film', 'U.S. release date',	'Director(s)'),
          ('Iron Man',	'May 2, 2008', 'Jon Favreau'),
          ('The Incredible Hulk', 'June 13, 2008', 'Louis Leterrier'),
          ('Thor', 'May 6, 2011', 'Kenneth Branagh'),
          ('Captain America: The First Avenger', 'July 22, 2011',	'Joe Johnston'),]
```

```
for film in marvel:
  print(film)
# It is not formatted properly
```

```
# enter variable space to align properly
for film in marvel:
  print(f'{film[0]:{40}}, {film[1]:{20}}, {film[2]:{10}}')
```

```
 # indicater between lines

# <   left align
# >   right align
# ^   centre align

for film in marvel:
  print(f'{film[0]:<{40}}, {film[1]:<{20}}, {film[2]:>{20}}')
```

```
# .  dot

for film in marvel:
  print(f'{film[0]:^{40}}, {film[1]:{20}}, {film[2]:.>{20}}')
```
