# 1) Reading PDFs

```
! pip install -q PyPDF2
```

```
import PyPDF2 as pdf
```

```
file = open('/content/sample.pdf',mode='rb')
# 'rb' for read binary
```

```
file
```

```
# create an object
pdf_reader = pdf.PdfFileReader(file)
```

```
pdf_reader
```

```
pdf_reader.numPages
```

```
page1 = pdf_reader.getPage(0)
```

```
print(page1)
# output is in binary form
```

```
# to get the text data
page1_text = page1.extractText()
```

```
print(page1_text)
```

```
page2 = pdf_reader.getPage(1)
```

```
# to get the text data
page2_text = page2.extractText()
```

```
print(page2_text)
```

```
# close the file
file.close()
```

# 2) Adding to PDFs

```
# We can not write to PDFs using Python,
# because of the differences between single string type of Python, variety of fonts, placements, and other parameters that a PDF could have
```

```
pdf_writer = pdf.PdfFileWriter()
```

```
pdf_writer.addPage(page1)
```

```
output = open('new_file.pdf',mode='wb')
# 'wb'=write binary
```

```
pdf_writer.write(output)
```

```
output.close()
```

```
# under the files, 'new_file' is created with the content of page1
# double click on the file to check the content
```
